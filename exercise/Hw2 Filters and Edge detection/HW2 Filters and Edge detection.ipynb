```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def apply_filter(image, kernel):
    """Applies a filter (kernel) to an image using convolution."""
    # Get dimensions
    image_h, image_w = image.shape
    kernel_h, kernel_w = kernel.shape

    # Determine padding size
    pad_h = kernel_h // 2
    pad_w = kernel_w // 2

    # Pad the image with zeros
    padded_image = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant', constant_values=0)
    plt.imshow(padded_image, cmap='gray')


    # Output image
    filtered_image = np.zeros_like(image)

    for i in range(image_h):
        for j in range(image_w):
            # Extract the region of interest (ROI) from the padded image
            region = padded_image[i:i+kernel_h, j:j+kernel_w]
        
            # Perform element-wise multiplication and sum the result
            filtered_image[i, j] = np.sum(region * kernel)



    # Clip values to valid range [0, 255]
    filtered_image = np.clip(filtered_image, 0, 255)

    return filtered_image
```

```python
image = cv2.imread('gates_2.jpg', cv2.IMREAD_GRAYSCALE)
kernel= np.array([[1, 1, 1],
                        [1, 1, 1],
                        [1, 1, 1]]) 
```

```python
apply_filter(image, kernel)
```

```python
plt.imshow(image, cmap='gray')
```

### Smoothing Filter

```python
# Load the image and convert to grayscale
image = cv2.imread('gates_2.jpg', 0)
image_array = np.array(image)
```

```python
# Define a blur kernel
blur_kernel = np.array([[1, 1, 1],
                        [1, 1, 1],
                        [1, 1, 1]]) / 9.0


# Apply the blur filter
blurred_image = apply_filter(image_array, blur_kernel)

# Display the result
plt.imshow(blurred_image, cmap='gray')
plt.axis('off')
plt.title("Blurred Image")
plt.show()
```

### Sharpening Filter

```python
# Define a sharpening kernel

#-------Todo-----------#
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])
# Apply the sharpening filter
sharpened_image = apply_filter(image_array, sharpen_kernel)

# Display the result
plt.imshow(sharpened_image, cmap='gray')
plt.axis('off')
plt.title("Sharpened Image")
plt.show()
```

### Edge Detection Filter

```python
#-------Todo-----------#
edge_kernel = np.array([[-1, -1, -1],
                        [-1, 8, -1],
                        [-1, -1, -1]])

# Apply the edge detection filter
edge_image = apply_filter(image_array, edge_kernel)

# Display the result
plt.imshow(edge_image, cmap='gray')
plt.axis('off')
plt.title("Edge Detection")
plt.show()
```
